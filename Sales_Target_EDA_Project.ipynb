```python
category = df["Category"]
target = df["Target"]

plt.figure(figsize=(10,10))
plt.barh(category, target, color = "skyblue")
plt.xlabel("Target")
plt.ylabel("Category")
plt.title("Category vs Target")
plt.grid()
plt.show()

#clothing and electronics have the highest targets
```

```python
pie1 = plt.pie(target, labels = category, autopct ="%0.1f%%", shadow = True, radius = 2)
pie2 =plt.pie([5], colors = "white", radius = 1)
plt.title("Targets per Category")
plt.show()
```

```python
sns.histplot(target, color = "darkgreen")
plt.grid()

#most of the sales targets lie between 9000 and 10000, 15000 and 16000
```

```python
df.corr()
```

```python
sns.violinplot(target, showmeans = True, showmedians = True, color = "Yellow")
plt.grid()
```

```python
# Intepretation of the violin plot above:

# 1. Maximum value is 16,000
# 2. upper quartile (Q3) = 14250
# 3. lower quartile (Q1) = 10000
# 4. Minimum value is 9000
# 5. the median number is 11750
```
